```python
df_22 = df.loc[df['Year'] == 2022]
df_22_est = df_22.loc[df_22['Category'] == 'Disability Estimates']
remove_cols = {'Any Disability', 'No Disability', 'Overall'}
mask = [d3 not in remove_cols for d3 in df_22_est['Stratification1']]
df_22_stratification = df_22_est.loc[mask]
for strat, group in df_22_stratification.groupby("Stratification1"):
    #print(strat)
    #print(group['Indicator'].iloc[0])
    loc_sum1, loc_sum2 = 0, 0
    for location, subgroup in group.groupby("LocationAbbr"):
        print(location)
        for response, subsubgroup in subgroup.groupby("Response"):
            data_cols = [col for col in subsubgroup.columns if 'data' in col.lower()]
            #print(subsubgroup[data_cols])
            if "male" in response.lower():
                try:
                    loc_sum1 += int(subsubgroup['Number'].iloc[0])
                except ValueError as e:
                    pass
                    #print(subsubgroup['WeightedNumber'].iloc[0])

            elif response in {"18-44", "45-64", "65+"}:
                try:
                    loc_sum2 += int(subsubgroup['Number'].iloc[0])
                except ValueError as e:
                    pass
                    #print(subsubgroup['WeightedNumber'].iloc[0])
    break
    if loc_sum1 != loc_sum2:
        print(f"Uneven: {loc_sum1}, {loc_sum2}, {loc_sum1 - loc_sum2}")
```
