Employment Levels

```python
def get_response_percentages(df, indicator):
    df_22 = df.loc[df['Year'] == 2022]
    df_22_est = df_22.loc[df_22['Indicator'] == indicator]
    output_df = pd.DataFrame(columns=['Disability_Type', 'Response_Value', 'Percentage'])
    remove_cols = {'Any Disability', 'No Disability', 'Overall'}
    mask = [d3 not in remove_cols for d3 in df_22_est['Stratification1']]
    df_22_stratification = df_22_est.loc[mask]
    for strat, group in df_22_stratification.groupby("Stratification1"):
        print(strat)
        totals = 0
        for i, (location, subgroup) in enumerate(group.groupby("LocationAbbr")):
            dist_dict = {}
            for response, subsubgroup in subgroup.groupby("Response"):
                data_cols = [col for col in subsubgroup.columns if 'data' in col.lower()]
                if response not in dist_dict:
                    dist_dict[response] = 0
                try:
                    if not all(subsubgroup['Data_Value_Footnote'].isna()):
                        pass
                    else:
                        dist_dict[response] += float(subsubgroup['Data_Value'].iloc[0])
                except:
                    pass
            if sum(dist_dict.values()) < 95:
                totals += 1
        print(sum(dist_dict.values()))
        dist_dict['Unknown'] = 100 - sum(dist_dict.values())
        print(dist_dict)
        for k, v in dist_dict.items():
            output_df.loc[len(output_df)] = [strat, k, v]
    return output_df
get_response_percentages(df, 'Smoking status among adults 18 years of age or older')
```
